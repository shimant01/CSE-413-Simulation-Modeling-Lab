```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Create Your Personal Matrix
Let:
  
    ● d1 = last digit of your student ID = 6
    
    ● d2 = second last digit of your student ID= 4

```python
A=np.array([[6+2,4+1],[2*6,4+2]])
print(A)
```

2. Analyze Matrix A
Using NumPy, compute:
    
      ● Shape of A
      
      ● Determinant
      
      ● Rank
      
      ● Eigenvalues
      
      ● Inverse (only if determinant ≠ 0)

```python
shape_of_A = A.shape
print("shape of A  (row,coloumn) =",shape_of_A)

det_A = np.linalg.det(A)
print("determinant of A :",det_A)

rank_A = np.linalg.matrix_rank(A)
print("rank of A :",rank_A)

eigenvalues,eigenvectors=np.linalg.eig(A)
print("eigenvalues of A:",eigenvalues)

if det_A !=0:
  inv_A = np.linalg.inv(A)
else:
  inv_A = "Matrix is singular, cannot compute inverse"
print("inverse of A :", inv_A)
```
