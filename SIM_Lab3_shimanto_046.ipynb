```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

Student's Assignment 1: Basic Matrix Operations
1. Create two 3Ã—3 matrices with random integers in Python.
2. Perform addition, subtraction, and multiplication operations on these matrices.
3. Compute the determinant, inverse, and rank of one of the matrices.

```python
#1
A = np.random.randint(1,13,(3,3))
B = np.random.randint(1,13,(3,3))
print("A:\n",A)
print("B:\n",B)
#2
print("addition of A and B :\n",A+B)
print("subtraction of A and B:\n",A-B)
print("multiplication of A and B\n",np.dot(A,B))
#3
det_A=np.linalg.det(A)
print("determinant of A:\n",det_A)
if det_A !=0:
  inv_A=np.linalg.inv(A)
else:
  inv_A="Can not compute inverse\n"
print("inverse matrix of A :\n",inv_A)
print("rank of A :",np.linalg.matrix_rank(A))
```

2. Generate and Visualize a Random Matrix:
  
  a. Create a 4x4 matrix with random values of your choice.
  
  b. Display the matrix using a heatmap with appropriate color mapping (e.g.,seaborn or matplotlib).
  
  c. Add labels to the heatmap to indicate row and column numbers.

```python
#a
matrix=np.random.randint(1,13,(4,4))
print("matrix:\n",matrix)

#b
sns.heatmap(matrix,annot=True,cmap='coolwarm',linewidths=0.5,linecolor='black')
plt.title('heatmap of 4x4 matrix')

#c
plt.xlabel('column')
plt.ylabel('rows')
plt.show()
```

3. Matrix Operations and Visualization:

 a. Create two 4x4 matrices with random values of your choice.

b. Perform the following operations:
i. Matrix Addition
ii. Matrix Subtraction
iii. Matrix Multiplication (ensure matrices are compatible for
multiplication)


c. For each of these operations, visualize the result using bar plots. Each bar
should represent one element of the matrix.

```python
#a
a=np.random.randint(1,13,(4,4))
print("matrix a:\n",a)
b=np.random.randint(1,13,(4,4))
print("matrix b:\n",b)

#b
add=a+b
sub=a-b
mul=np.dot(a,b)
print("matrix addition of a,b:\n",add)
print("matrix subtraction of a,b\n",sub)
print("matrix multiplication of a,b\n",mul)

#c
fig,axes = plt.subplots(1,3,figsize=(25,5))
#sum
axes[0].bar(range(1,17), add.flatten(), color='g')
axes[0].set_title('matrix addition')
axes[0].set_xlabel('index')
axes[0].set_ylabel('value')
#sub
axes[1].bar(range(1,17), sub.flatten(), color='r')
axes[1].set_title('matrix substraction')
axes[1].set_xlabel('index')
axes[1].set_ylabel('value')
#mul
axes[2].bar(range(1,17), mul.flatten(), color='b')
axes[2].set_title('matrix multiplication')
axes[2].set_xlabel('index')
axes[2].set_ylabel('value')
plt.tight_layout()
plt.show()
```
