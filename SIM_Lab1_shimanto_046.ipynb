```python
import numpy as nd
from sympy import Matrix
import matplotlib.pyplot as plt
import  seaborn as sns

mat=np.array([[1,3],[4,6],[2,2]])
#Trigonometric functions
sin_val=np.sin(mat)
cos_val=np.cos(mat)
tan_val=np.tan(mat)
asin_val=np.arcsin(np.clip(mat/np.max(mat),-1,1))
acos_val=np.arccos(np.clip(mat/np.max(mat),-1,1))
atan_val=np.arctan(np.clip(mat/np.max(mat),-1,1))

#Exponential & logarithmic functions
exp_val=np.exp(mat)
log_val=np.log(mat)

#Absolute value & square root
abs_val=np.abs(mat)
sqrt_val=np.sqrt(mat)

#Remainder when divided by 2
rem_val=np.remainder(mat,2)

#Rounding (round, floor, ceil)
round_val=np.round(mat)
floor_val=np.floor(mat)
ceil_val=np.ceil(mat)

# Display results
print("Matrix:\n", mat)
print("\nsin:\n", sin_val)
print("\ncos:\n", cos_val)
print("\ntan:\n", tan_val)
print("\nasine:\n", asin_val)
print("\nacosine:\n", acos_val)
print("\natangent:\n", atan_val)
print("\nexp:\n", exp_val)
print("\nlog (natural):\n", log_val)
print("\nabs:\n", abs_val)
print("\nsqrt:\n", sqrt_val)
print("\nrem (remainder when divided by 2):\n", rem_val)
print("\nround:\n", round_val)
print("\nfloor:\n", floor_val)
print("\nceil:\n", ceil_val)
```

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

vec=np.array([[1.3,4.6,2.2],[5.0,3.5,4.4]])

#Max & Min (with indices)
max_val=np.max(vec)
max_index=np.argmax(vec)
min_val=np.min(vec)
min_index=np.argmin(vec)

#Vector length
vec_length=len(vec)

#Sorted version (both row-wise descending and entire matrix descending)
sorted_vec=np.sort(vec,axis=1)
flatten_vec = np.sort(vec.flatten())

#Sum, Product
sum_val=np.sum(vec)
prod_val=np.prod(vec)

#Median, Mean, Standard Deviation
median_val=np.median(vec)
mean_val=np.mean(vec)
std_dev=np.std(vec)

# Display results
print("Max value:", max_val, "at index", max_index)
print("Min value:", min_val, "at index", min_index)
print("Length of vector:", vec_length)
print("Sorted vector:", sorted_vec)
print("Sum of elements:", sum_val)
print("Product of elements:", prod_val)
print("Median value:", median_val)
print("Mean value:", mean_val)
print("Standard deviation:", std_dev)
```
